# Predicting AirBnB Price Prediction

One of the biggest struggles for a property owner looking to rent their unit is determining what price they should list their unit for. Overestimating this value can be risky as renters may not be willing to pay such a high price, leading to less or no business. Underestimating this value might increase business, but minimize profit and perhaps make it not worth renting in the first place. And for these exact reasons, it is not worth it for property owners to experiment with prices until they find the correct one. 

Instead, why not base rental prices on other similar units? And we do not mean asking renting neighbours what their listing price is. With our neural network model for predicting AirBnB prices, we can estimate exactly what price a property owner should list their unit for based on 59 features of the space ranging from location to number of bathrooms to cleaning fees. 

To develop our neural network, we used a data set of AirBnB listings in Austin, TX. The original data set consisted of 11,339 listing with 63 features each. The data was incomplete, with missing values in many columns, as well as some undesirable formatting. So, our model development started with cleaning up the data. 

But first, we must import our packages. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import explained_variance_score, mean_squared_error, r2_score
import tensorflow as tf;
from tensorflow import keras
from tensorflow.keras import layers
from keras import models, layers, optimizers, regularizers
from keras.utils.vis_utils import model_to_dot
from IPython.display import SVG
```

By simply looking at the data, we can see that a number of columns are not relevant to our model. Rental prices are not going to vary based on the host's ID number in the AirBnB database. So, we drop these columns.

```python
#DATA READING
data = pd.read_csv("../data/reduced_listings.csv", low_memory = False)
print("total shape initially: ", data.shape)

# DATA CLEANING
data = pd.DataFrame(data.drop(["notes", "transit", "access", "host_name", "host_id", "name", \
                  "host_since", "calendar_last_scraped", "host_location", "host_acceptance_rate", \
                 "host_verifications", "host_total_listings_count", "neighbourhood_group_cleansed", \
                "host_neighbourhood", "amenities", "city", "state", "country_code", "smart_location", \
                "is_location_exact", "calendar_updated", "license", "jurisdiction_names", "weekly_price", \
               "monthly_price", "host_response_time", "host_response_rate", "square_feet", "security_deposit"] \
                , axis = 1))

print("size after dropping: ", data.shape)
```

Next we replace some of the features to make them numbers instead of unnecessary strings. All of the prices in the data set are formatted with dollar signs and commas, which are not helpful in the processing of the data. Additionally, boolean data is currently formatted as 't' and 'f', which we convert to binary. 

Finally, note at the bottom of this cell how we exclude data points where the price exceeds 2000. This is because these data points are outliers, and the vast majority of the data falls below this threshold. When the data is included, it skews the model. 

```python
data["cleaning_fee"] = data["cleaning_fee"].str.replace(',', '')
data["cleaning_fee"] = data["cleaning_fee"].str.replace('$', '')
data["cleaning_fee"] = data["cleaning_fee"].str.replace(' ', '')
data["cleaning_fee"] = data["cleaning_fee"].astype(float)

data["extra_people"] = data["extra_people"].str.replace(',', '')
data["extra_people"] = data["extra_people"].str.replace('$', '')
data["extra_people"] = data["extra_people"].str.replace(' ', '')
data["extra_people"] = data["extra_people"].astype(float)

data["host_identity_verified"] = data["host_identity_verified"].str.replace('t', '1')
data["host_identity_verified"] = data["host_identity_verified"].str.replace('f', '0')
data["host_identity_verified"] = data["host_identity_verified"].astype(float)

data["host_is_superhost"] = data["host_is_superhost"].str.replace('t', '1')
data["host_is_superhost"] = data["host_is_superhost"].str.replace('f', '0')
data["host_is_superhost"] = data["host_is_superhost"].astype(float)

data["requires_license"] = data["requires_license"].str.replace('t', '1')
data["requires_license"] = data["requires_license"].str.replace('f', '0')
data["requires_license"] = data["requires_license"].astype(float)

data["price"] = data["price"].str.replace(',', '')
data["price"] = data["price"].str.replace('$', '')
data["price"] = data["price"].str.replace(' ', '')
data["price"] = data["price"].astype(float)

data = data[data.price < 2000]
```

We also want to separate our target values from the rest of the data. 

```python
price = pd.DataFrame(data["price"])
print("price size: ", price.shape)

data_no_nan = data.drop("price", axis = 1)
```

Missing values in the data are an issue, so we must find a way to deal with them. First we should understand how frequently data is missing. 

```python
print(data_no_nan.isna().sum())
```

As we see from the number of missing cells from each column, there is a large amount of missing data. While one possible solution would be to delete any rows with missing cells, this would lead to deletion of a minimum of 2500+ rows. Since this is almost a quarter of our data set, that is not an acceptable option.

Instead, we look at the two different types of data we have: numbers and strings. For number columns, we can replace the empty data with the median value for that column. This may not be a perfect solution, but it allows us to keep the data and make our best guess at what it would have been. For columns with strings, we replace the empty data with the most common values in that column. 

```python
#DEAL WITH MISSING VALUES
data_no_nan.dropna(inplace = True, how = "all")

for i in data_no_nan.columns:
    if data_no_nan[i].dtype == int or data_no_nan[i].dtype == float:
        data_no_nan.replace('', data_no_nan[i].median(), inplace = True)
        data_no_nan.replace(np.nan, data_no_nan[i].median(), inplace = True)
    else:
        data_no_nan.replace('', data_no_nan[i].mode(), inplace = True)

print(data_no_nan.isna().sum())
```

Next we deal with the issue of columns with string values. We cannot use this data directly in our neural network, so instead we must create new columns that act as dummy variables for all the different values in a categorical column. Each of these dummy variables take on a value of 1 or 0. This introduces many new columns to the data, but makes it usable in a neural network. 

```python
#ONE HOT ENCODING
columns = ['host_has_profile_pic', 'street', 'zipcode', \
           'market', 'country', 'property_type', 'room_type','bed_type', \
           'has_availability', 'first_review', 'last_review', 'instant_bookable', \
           'is_business_travel_ready', 'cancellation_policy', 'require_guest_profile_picture', \
           'require_guest_phone_verification', 'neighbourhood']

data_with_dummies = pd.get_dummies(data = data_no_nan, columns = columns)
print("new data size: ", data_with_dummies.shape)
```

We perform one additional simple feature selection method, using a variance threshold to ensure that none of the columns have all the same value, since this would not help with the model. There are only 4 of these columns in our data set. 

```python
#FEATURE SELECTION
print("initial shape: ", data_with_dummies.shape)
feature_selection = VarianceThreshold()
data_with_dummies = feature_selection.fit_transform(data_with_dummies)
feature_mask = feature_selection.get_support()
del_features = np.array([i if feature_mask[i] == False else -1 \
                        for i in range(feature_mask.shape[0])])
del_features = del_features[del_features != -1]
print("final shape: ", data_with_dummies.shape)
```

To finalize the formatting of the data, we split it into training data and testing data. Here, we save 20% of the data for testing. After splitting the data into these categories, we normalize it. At this point, we can begin to build the model. 

```python
#SPLITTING
x_train, x_test, y_train, y_test = train_test_split(data_with_dummies, price, \
                                                        test_size = 0.2)

print("x train shape: ", x_train.shape)
print("x test shape: ", x_test.shape)
print("y train shape: ", y_train.shape)
print("y test shape: ", y_test.shape)

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
y_train = scaler.fit_transform(y_train)
y_test = scaler.fit_transform(y_test)

#x_train = pd.DataFrame(normalize(x_train, axis = 1))
#x_test = pd.DataFrame(normalize(x_test, axis = 1))
#y_train = pd.DataFrame(normalize(y_train, axis = 0))
#y_test = pd.DataFrame(normalize(y_test, axis = 0))


# print("y_train info: ", y_train.describe())
# print("x_train info: ", x_train.describe())
```

When constructing our model we have a number of choices to make. Some of the most important decisions are
- how many hidden layers our neural network should use
- how many nodes each layer should have
- which activation function a later should use
- which loss function to use
- how many epochs to run

While we are using mean squared error for our loss function and relu (a linear function) for our activation function, we found it worth exploring how to vary the other options.

For our first run-through, we use a model with four layers and 128, 256, 256, and 1 nodes repectively. We run 50 epochs for this model.

```python
# BUILDING THE MODEL
nn2 = models.Sequential()
nn2.add(layers.Dense(128, input_shape=(x_train.shape[1],), activation='relu'))
nn2.add(layers.Dense(256, activation='relu'))
nn2.add(layers.Dense(256, activation='relu'))
nn2.add(layers.Dense(1, activation='relu'))

# Compiling the model
nn2.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['mean_squared_error'])

# Printing the model summary
print(nn2.summary())

# Visualising the neural network
SVG(model_to_dot(nn2, show_layer_names=False, show_shapes=True).create(prog='dot', format='svg'))

# Training the model
nn2_history = nn2.fit(x_train,
                  y_train,
                  epochs=50,
                  batch_size=256,
                  validation_split = 0.1)
```

```python
def nn_model_evaluation(model, skip_epochs=0, x_train=x_train, x_test=x_test, y_train=y_train, y_test=y_test):
    """
    For a given neural network model that has already been fit, prints for the train and tests sets the MSE and r squared
    values, a line graph of the loss in each epoch, and a scatterplot of predicted vs. actual values with a line
    representing where predicted = actual values. Optionally, a value for skip_epoch can be provided, which skips that
    number of epochs in the line graph of losses (useful in cases where the loss in the first epoch is orders of magnitude
    larger than subsequent epochs). Training and test sets can also optionally be specified.
    """

    # MSE and r squared values
    y_test_pred = model.predict(x_test)
    y_train_pred = model.predict(x_train)
    
    y_test_pred_unscaled = scaler.inverse_transform(y_test_pred)
    y_train_pred_unscaled = scaler.inverse_transform(y_train_pred)
    y_train_unscaled = scaler.inverse_transform(y_train)
    y_test_unscaled = scaler.inverse_transform(y_test)
    
    print("Training MSE:", round(mean_squared_error(y_train, y_train_pred),4))
    print("Validation MSE:", round(mean_squared_error(y_test, y_test_pred),4))
    print("\nTraining r2:", round(r2_score(y_train_unscaled, y_train_pred_unscaled),4))
    print("Validation r2:", round(r2_score(y_test_unscaled, y_test_pred_unscaled),4))
    
    # Line graph of losses
    model_results = model.history.history
    plt.plot(list(range((skip_epochs+1),len(model_results['loss'])+1)), model_results['loss'][skip_epochs:], label='Train')
    plt.plot(list(range((skip_epochs+1),len(model_results['val_loss'])+1)), model_results['val_loss'][skip_epochs:], label='Test', color='green')
    plt.legend()
    plt.title('Training and test loss at each epoch', fontsize=14)
    plt.show()
    
    # Scatterplot of predicted vs. actual values
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    fig.suptitle('Predicted vs. actual values', fontsize=14, y=1)
    plt.subplots_adjust(top=0.93, wspace=0)
    
    ax1.scatter(y_test_unscaled, y_test_pred_unscaled, s=2, alpha=0.7)
    ax1.plot(list(range(0,2000)), list(range(0,2000)), color='black', linestyle='--')
    ax1.set_title('Test set')
    ax1.set_xlabel('Actual values')
    ax1.set_ylabel('Predicted values')
    
    ax2.scatter(y_train_unscaled, y_train_pred_unscaled, s=2, alpha=0.7)
    ax2.plot(list(range(0,2000)), list(range(0,2000)), color='black', linestyle='--')
    ax2.set_title('Train set')
    ax2.set_xlabel('Actual values')
    ax2.set_ylabel('')
    ax2.set_yticklabels(labels='')
    
    plt.show()
```

```python
nn_model_evaluation(nn2)
```

As we can see, the MSE for our first model is 0.0102 with a validation r squared value of 0.4522. Not bad! But let's explore some other models to compare them.

In our next model we introduce early stopping so as to make the model stop when training and validation loss converge. Since we can see that the lines converge early, we will only use 5 epochs. In addition, we will introduce two more hidden layers.

```python
# BUILDING THE MODEL
nn3 = models.Sequential()
nn3.add(layers.Dense(128, input_shape=(x_train.shape[1],), activation='relu'))
nn3.add(layers.Dense(256, activation='relu'))
nn3.add(layers.Dense(256, activation='relu'))
nn3.add(layers.Dense(256, activation='relu'))
nn3.add(layers.Dense(256, activation='relu'))
nn3.add(layers.Dense(1, activation='relu'))


# Compiling the model
nn3.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['mean_squared_error'])

# Printing the model summary
print(nn3.summary())

# Visualising the neural network
SVG(model_to_dot(nn3, show_layer_names=False, show_shapes=True).create(prog='dot', format='svg'))

# Training the model
nn3_history = nn3.fit(x_train,
                  y_train,
                  epochs=5,
                  batch_size=256,
                  validation_split = 0.1)

nn_model_evaluation(nn3)
```

This model performs slightly better. The MSE only increases to 0.0093 despite the r squared value increasing to 0.5028.

Let's try with fewer layers instead.

```python
# BUILDING THE MODEL
nn4 = models.Sequential()
nn4.add(layers.Dense(128, input_shape=(x_train.shape[1],), activation='relu'))
nn4.add(layers.Dense(256, activation='relu'))
nn4.add(layers.Dense(1, activation='relu'))

# Compiling the model
nn4.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['mean_squared_error'])

# Printing the model summary
print(nn4.summary())

# Visualising the neural network
SVG(model_to_dot(nn4, show_layer_names=False, show_shapes=True).create(prog='dot', format='svg'))

# Training the model
nn4_history = nn4.fit(x_train,
                  y_train,
                  epochs=5,
                  batch_size=256,
                  validation_split = 0.1)

nn_model_evaluation(nn4)
```

This model does not get better. The MSE is 0.0102 and the r squared value is 0.4546.

Perhaps more hidden layers like before will help, but this time with the addition of more nodes in the hidden layers.

```python
# BUILDING THE MODEL
nn5 = models.Sequential()
nn5.add(layers.Dense(128, input_shape=(x_train.shape[1],), activation='relu'))
nn5.add(layers.Dense(256, activation='relu'))
nn5.add(layers.Dense(256, activation='relu'))
nn5.add(layers.Dense(512, activation='relu'))
nn5.add(layers.Dense(512, activation='relu'))
nn5.add(layers.Dense(1, activation='relu'))

# Compiling the model
nn5.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['mean_squared_error'])

# Printing the model summary
print(nn5.summary())

# Visualising the neural network
SVG(model_to_dot(nn5, show_layer_names=False, show_shapes=True).create(prog='dot', format='svg'))

# Training the model
nn5_history = nn5.fit(x_train,
                  y_train,
                  epochs=5,
                  batch_size=256,
                  validation_split = 0.1)

nn_model_evaluation(nn5)
```

This model improves upon the previous ones with a MSE of 0.0098 and a r squared value of 0.4738.

Maybe even more will help.

```python
# BUILDING THE MODEL
nn6 = models.Sequential()
nn6.add(layers.Dense(128, input_shape=(x_train.shape[1],), activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(256, activation='relu'))
nn6.add(layers.Dense(1, activation='relu'))

# Compiling the model
nn6.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['mean_squared_error'])

# Printing the model summary
print(nn6.summary())

# Visualising the neural network
SVG(model_to_dot(nn6, show_layer_names=False, show_shapes=True).create(prog='dot', format='svg'))

# Training the model
nn6_history = nn6.fit(x_train,
                  y_train,
                  epochs=5,
                  batch_size=256,
                  validation_split = 0.1)

nn_model_evaluation(nn6)
```

This model is slightly worse than the last, suggesting that adding more layers does not necessarily help the model.

In conclusion, the best models are nn3 and nn5, with nn3 scoring slightly better on the r squared value and slightly better on the MSE. nn3, being the better of the two, used 6 hidden layers and only 5 epochs. The model performs very well at predicting a difficult set of data.
